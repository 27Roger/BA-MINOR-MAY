```python
#Preprocessing the Data
# Resize
#Flatten
import os
import matplotlib.pyplot as plt
import numpy as np

from skimage.io import imread
from skimage.transform import resize

target_output =[]
images = [] #2D
flat_data = []

imgpath = '/content/images'
categories = ['single rose jpg','single sunflower jpg','single white lily jpg']

for category in categories:
  class_num = categories.index(category)
  path = os.path.join(imgpath,category)
  print(path)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    #plt.imshow(img_array)
    #break
#break
    img_resize = resize(img_array,(150,150,3))#Normalizes values from 0 to 1
    flat_data.append(img_resize.flatten())
    images.append(img_resize)
    target_output.append(class_num)

flat_data = np.array(flat_data)
target_output = np.array(target_output)
images = np.array(images)
```

```python
#Split data into Training and testing 

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(flat_data, target_output, test_size = 0.3, random_state = 109)
```

```python
import pandas as pd
df = pd.DataFrame(flat_data)
df['Target'] = target_output
df
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
]

svc = svm.SVC(probability = True)
clf = GridSearchCV(svc, param_grid)
clf.fit(xtrain,ytrain)
```

```python
ypred = clf.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(ypred,ytest)
```

```python
confusion_matrix(ypred,ytest)
```

```python
#Save model using Pickle lib
import pickle
pickle.dump(clf, open('img_model.p','wb'))
```
